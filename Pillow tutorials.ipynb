## Transposing a image
 Transposing means changing the orientation of the image

```python
tr1_img = img1.transpose(Image.FLIP_LEFT_RIGHT)
tr2_img = img1.transpose(Image.FLIP_TOP_BOTTOM)
tr3_img = img1.transpose(Image.ROTATE_90)
tr4_img = img1.transpose(Image.ROTATE_180)


pl.imshow(tr3_img)
```

## Resizing
Thsi func hepls to change the size of a img wrt the height and the width

There are six methods used for resizing


**BILINEAR,NEAREST,BOX,HAMMING,BICUBIC,LANCZOS,ANTIALIAS**


```python
newsize=(300,300)

r_img1=img1.resize((300,300),Image.BILINEAR)
#both are same
r_img2=img1.resize(newsize,Image.NEAREST)
r_img3=img1.resize(newsize,Image.BOX)
r_img4=img1.resize(newsize,Image.HAMMING)
r_img5=img1.resize(newsize,Image.ANTIALIAS)
r_img6=img1.resize(newsize,Image.BICUBIC)
r_img6=img1.resize(newsize,Image.LANCZOS)
pl.imshow(r_img5)
```

## IMAGE ENHANCEMENT 
There are different types of image enhancement techniques which help to increase the pixel quality so that the image may be more suitable to be used .
they are **Colour ,Constarst,Sharness and Brightness Enhancement**

```python
# First importing the module
from PIL import ImageEnhance

enh_img = img1.copy()
enh_img1 = ImageEnhance.Color(enh_img).enhance(0.05)
enh_img2 = ImageEnhance.Contrast(enh_img).enhance(2.5)
enh_img3 = ImageEnhance.Brightness(enh_img).enhance(1.5)
enh_img4 = ImageEnhance.Sharpness(enh_img).enhance(2)

enh_img1.show()
```

## Alpha Blending
Blending means combining two images together .**For blending to occur both the images should be of the same size**

```python
ib1 = img1.copy()
ib2 = img2.copy()

ib1 = ib1.resize(ib2.size)

imageblend = Image.blend(ib1,ib2,0.25)
pl.imshow(imageblend)
```

## Flipping the channels 
**Very Important for Image processing**


```python
img_chnl=img1.copy()
# splitting the channel
r,g,b = img_chnl.split()
pl.imshow(r)
```

## To merge those flipped channels together in opp order

```python
# Merging the channel
im = Image.merge("RGB",(b , g , r))
pl.imshow(im)
```

```python
img3 = Image.open('M:\py_prgms\Baboon.jpg')
ar3=np.array(img3)
img3.show()
```

```python
def get_concat(im1, im2):
    #https://note.nkmk.me/en/python-pillow-concat-images/
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst
```

```python
r , g , b = img3.split()
get_concat(img3, r)
```
